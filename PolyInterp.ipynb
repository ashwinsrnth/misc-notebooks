## Polynomial interpolation with a single polynomial of high-degree

* Constructing a singleinterpolating polynomial for the points $(x_i, y_i), i = 1,...,n$, for large $n$ is generally considered a bad idea. 

* Consider *Runge's function* here:

```python
%pylab inline
def runge(x):
    return 1./(1 + x**2)

x = linspace(-5, 5)
plot(x, runge(x))
text(0.9, 0.9, r'$\frac{1}{1+x^2}$', fontsize=20)
title('Runge function')
```

* Intuition suggest that the accuracy of the interpolating polynomial increases as the degree (and the number of points) increases.

* For this function (and many others), accuracy gets *worse* as the degree increases, due to overshoot, especially near the endpoints. The following plots illustrate this:

```python
%pylab inline


figure(figsize(11,4))

x = linspace(-5, 5, 100)
px = linspace(-5, 5, 5)
py = runge(px)
p = poly1d(polyfit(px, py, 4))

subplot(121)
plot(px, py, 'o', label='Points for interpolation')
plot(x, p(x), '--', label='Polynomial curve')
plot(x, runge(x), label='Original curve')
title('5 uniformly spaced points')
legend(loc='best', prop={'size':9})

x = linspace(-5, 5, 100)
px = linspace(-5, 5, 9)
py = runge(px)
p = poly1d(polyfit(px, py, 8))

subplot(122)
plot(px, py, 'o', label='Points for interpolation')
plot(x, p(x), '--', label='Polynomial curve')
title('9 uniformly spaced poitns')
plot(x, runge(x), label='Original curve')
legend(loc='best', prop={'size':9})

tight_layout()
```

## Piecewise Polynomial Interpolation


* Given a set of points $P = {(x_i, y_i), i = 1, ... n}, with $x_1 < x_2 ... < x_n$, *piecewise polynomial interpolation* generally means

    - Several low degree polynomials, each interpolating a subset of $P$, are used rather than one $n-1$th degree polynomial
    - The breakpoints at which the polynomials change from one to another are called *knots*

### Advantages and disadvantages of continuous piecewise polynomials

* Less overshoot than high degree polynomials
* Easy to form and evaluate (e.g., use Lagrange form)
* Possible disadvantage - lack of smoothness at knots

#### Example: Continuous piecewise linear polynomial

* There is *knot* at each x_i
* There are $n-1$ pieces
* If $x$ is in $[x_i, x_i+1]$, then

$$ p(x) = p_i(x) = y_i + (x - x_i)\frac{y_{i+1} - y_i}{x_{i+1} - x_i} $$


```python
%pylab inline
from scipy.interpolate import PiecewisePolynomial

xi = arange(10)
yi = [1, 2, 5, 6, 8, 8, 7, 5, 4, 1]
plot(xi, yi, 'o-')
title('Piecewise continuous linear polynomial')
axis([0, 9, 0, 9])
```

## Spline Interpolation


**General Definition**: A piecewise polynomial of degree $k$ (i.e., each *piece* is a degree $k$ polynomial) is a *spline* if it has $k-1$ continuous derivatives *everywhere*, including at the knots

* The piecewise continuous linear polynomial described above is an example of a spline of degree 1.
* In general, a piecewise continuous polynomial of degree $k \geq 2$ is not a spline.
* Splines of such higher degrees can be constructed by imposing the required continuity constraints.
* The most common spline is the $cubic$ spline ($k = 3$)


For a given set of points $(x_i, y_i), i = 1...N$, a cubic spline interpolant has the form

$ p(x) = p_i(x) = {c_{0}}^{(i)} + {c_{1}}^{(i)}x + {c_{2}}^{(i)}x^2 + {c_{3}}^{(i)}x^3 $ 

for $x$ in $[x_i, x_{i+1}]$

So there are $4(n-1)$ coefficients to determine, i.e., $4(n-1)$ *degrees of freedom*

```python
%pylab inline
from scipy.interpolate import interp1d

xi = arange(10)
yi = [1, 2, 5, 6, 8, 8, 7, 5, 4, 1]
spline_fit = interp1d(xi, yi, kind='cubic')
x = linspace(0, 9, 100)

plot(xi, yi, 'o', label='Data points')
plot(x, spline_fit(x), label='Cubic spline')
title('Cubic spline interpolation')
```
